```python
# --- 1.1.4 Distribution of 'duration' (Needs Cleaning/Splitting First!) ---
# First, let's separate movies and TV shows to analyze their durations separately
movies_df = df[df['type'] == 'Movie'].copy() # Use .copy() to avoid SettingWithCopyWarning
tv_shows_df = df[df['type'] == 'TV Show'].copy()

# Convert movie duration to integer (in minutes)
# Example: '93 min' -> 93
movies_df['duration_minutes'] = movies_df['duration'].str.extract('(\d+)').astype(float)

# Convert TV show duration to integer (number of seasons)
# Example: '2 Seasons' -> 2
tv_shows_df['num_seasons'] = tv_shows_df['duration'].str.extract('(\d+)').astype(float)


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.histplot(data=movies_df, x='duration_minutes', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations (minutes)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.countplot(data=tv_shows_df, x='num_seasons', palette='rocket')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')
plt.tight_layout() # Adjusts plot parameters for a tight layout
plt.show()

print("\n--- Summary Statistics for Movie Durations ---")
print(movies_df['duration_minutes'].describe())
print("\n--- Value Counts for TV Show Seasons ---")
print(tv_shows_df['num_seasons'].value_counts().sort_index())

```

```python
# --- 1.1.5 Top 'country'ies (Requires Handling Multiple Countries) ---
# Create a list of all individual countries
countries_list = df['country'].dropna().str.split(', ').explode()

# --- FIX: Reset the index of countries_list ---
countries_list = countries_list.reset_index(drop=True)

plt.figure(figsize=(12, 7))
sns.countplot(y=countries_list, order=countries_list.value_counts().head(10).index, palette='crest')
plt.title('Top 10 Content Producing Countries on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()

print("\n--- Top 10 Countries by Content Count ---")
print(countries_list.value_counts().head(10))


# --- 1.1.6 Most common 'listed_in' (genres - Requires Handling Multiple Genres) ---
# Create a list of all individual genres
genres_list = df['listed_in'].dropna().str.split(', ').explode()

# --- FIX: Reset the index of genres_list ---
genres_list = genres_list.reset_index(drop=True)

plt.figure(figsize=(12, 7))
sns.countplot(y=genres_list, order=genres_list.value_counts().head(10).index, palette='cubehelix')
plt.title('Top 10 Most Common Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()

print("\n--- Top 10 Genres by Content Count ---")
print(genres_list.value_counts().head(10))
```

```python
#Biivariate Analysis
# --- 1.2.1 Content Type vs. Release Year ---
plt.figure(figsize=(15, 7)) 
sns.boxplot(data=df, x='release_year', y='type', palette='Set2',showfliers=False)
plt.title('Content Type vs. Release Year')
plt.xlabel('Release Year')
plt.ylabel('Content Type')  
plt.xticks(rotation=45)
plt.show()

# --- 1.2.2 Content Type vs. Rating ---
plt.figure(figsize=(12, 6)) 
sns.countplot(data=df, x='rating', hue='type', order=df['rating'].value_counts().index, palette='mako')
plt.title('Content Type vs. Rating')
plt.xlabel('Content Rating')
plt.ylabel('Number of Titles')  
plt.xticks(rotation=45)
plt.legend(title='Content Type')
plt.show()

# --- 1.2.3 Release Year vs. Duration ---
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.scatterplot(data=movies_df, x='release_year', y='duration_minutes', palette='coolwarm', s=50, alpha=0.6)
plt.title('Release Year vs. Movie Duration')   
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')    
plt.xticks(rotation=45,horizontalalignment='right')
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.regplot(data=tv_shows_df, x='release_year', y='num_seasons', scatter_kws={'s': 50, 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Release Year vs. TV Show Seasons')   
plt.xlabel('Release Year')  
plt.ylabel('Number of Seasons')
plt.xticks(rotation=45, horizontalalignment='right')
plt.tight_layout() # Adjusts plot parameters for a tight layout
plt.show()


```

```python
#Multivariate Analysis
# Three or More Variables - Often Visualized
# --- 1.3.1 Content Type, Rating, and Release Year ---
plt.figure(figsize=(15, 7)) 
sns.scatterplot(data=df, x='release_year', y='rating', hue='type', style='type', palette='Set1', s=100, alpha=0.7)
plt.title('Content Type, Rating, and Release Year')
plt.xlabel('Release Year')
plt.ylabel('Content Rating')
plt.xticks(rotation=45, horizontalalignment='right')
plt.legend(title='Content Type')
plt.show()

# --- 1.3.2 Bubble Chart: Movie Duration, Release Year, and Rating (A more complex visualization) ---
# Select a subset of data for clarity, e.g., top 5 ratings
top_5_ratings = movies_df['rating'].value_counts().head(5).index
movies_filtered_by_rating = movies_df[movies_df['rating'].isin(top_5_ratings)]

plt.figure(figsize=(15, 8))
sns.scatterplot(data=movies_filtered_by_rating, x='release_year', y='duration_minutes',
                hue='rating', size='duration_minutes', sizes=(20, 1000), alpha=0.6,
                palette='tab10') # Size of bubbles based on duration
plt.title('Movie Duration, Release Year, and Rating')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Ensure 'df' is loaded (from Step 0)

# --- 2.1 Convert 'date_added' to datetime objects and extract components ---
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce') # 'coerce' turns invalid dates into NaT (Not a Time)

# Drop rows where 'date_added' could not be parsed or was missing
df.dropna(subset=['date_added'], inplace=True)

df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
df['added_day_of_week'] = df['date_added'].dt.day_name()
df['added_quarter'] = df['date_added'].dt.quarter

print("\n--- New Date-related Features Added ---")
print(df[['date_added', 'added_year', 'added_month', 'added_day_of_week', 'added_quarter']].head())


# --- 2.2 Number of Directors/Actors/Countries/Genres ---

# For director, cast, country, listed_in - count the number of entries
# Fill NaN with empty string so split doesn't fail
df['num_directors'] = df['director'].fillna('').apply(lambda x: len(x.split(', ')) if x else 0)
df['num_cast'] = df['cast'].fillna('').apply(lambda x: len(x.split(', ')) if x else 0)
df['num_countries'] = df['country'].fillna('').apply(lambda x: len(x.split(', ')) if x else 0)
df['num_genres'] = df['listed_in'].fillna('').apply(lambda x: len(x.split(', ')) if x else 0)

print("\n--- New Count Features Added ---")
print(df[['title', 'director', 'num_directors', 'cast', 'num_cast',
          'country', 'num_countries', 'listed_in', 'num_genres']].head())

# Re-create movie and TV show specific duration columns after potential df modifications
movies_df = df[df['type'] == 'Movie'].copy()
tv_shows_df = df[df['type'] == 'TV Show'].copy()

# Ensure these columns are floats for calculations
movies_df['duration_minutes'] = movies_df['duration'].str.extract('(\d+)').astype(float)
tv_shows_df['num_seasons'] = tv_shows_df['duration'].str.extract('(\d+)').astype(float)

# Drop original 'duration' column if you only want the numeric ones
# df.drop('duration', axis=1, inplace=True) # Decide if you want to keep original string duration or drop it

print("\n--- New Duration Features Added ---")
```

```python
# --- 4.1.1 Number of titles added per year ---
plt.figure(figsize=(12, 7))
df['added_year'].value_counts().sort_index().plot(kind='line', marker='o', color='darkgreen')
plt.title('Number of Titles Added to Netflix Per Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# --- 4.1.2 Number of titles added per month (overall trend) ---
plt.figure(figsize=(12, 7))
# Map month numbers to names for better readability
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
df['added_month_name'] = df['added_month'].map(month_names)

sns.countplot(data=df, x='added_month_name', order=month_names.values(), palette='cool')
plt.title('Number of Titles Added to Netflix Per Month (Overall)')
plt.xlabel('Month Added')
plt.ylabel('Number of Titles')
plt.show()

# --- 4.1.3 Lag between release_year and date_added ---
# Calculate the lag (difference in years)
# Ensure release_year is numeric and handle potential missing values
df_lag = df.copy()
df_lag['release_year'] = pd.to_numeric(df_lag['release_year'], errors='coerce')
df_lag.dropna(subset=['release_year', 'added_year'], inplace=True)
df_lag['years_on_shelf'] = df_lag['added_year'] - df_lag['release_year']

plt.figure(figsize=(12, 7))
sns.histplot(data=df_lag[df_lag['years_on_shelf'] >= 0], # Focus on non-negative lags
             x='years_on_shelf', bins=30, kde=True, color='purple')
plt.title('Distribution of Years Between Content Release and Addition to Netflix')
plt.xlabel('Years on Shelf (Date Added - Release Year)')
plt.ylabel('Number of Titles')
plt.xlim(0, 50) # Limit x-axis for better readability, as some titles might be very old
plt.show()

print("\n--- Summary Statistics for Years on Shelf ---")
print(df_lag[df_lag['years_on_shelf'] >= 0]['years_on_shelf'].describe())
```
